```python
!pip install google.generativeai
import google.generativeai as genai
from tabulate import tabulate

# Configurar a API key - substitua 'YOUR_API_KEY' com sua chave da API do Google Generative AI
genai.configure(api_key='AIzaSyB7PW3dPzZJ90ec0wJK1aE3Eg_XU8cvpqg')

def training_generator():
    """Função principal do programa Personal Trainer IA."""

    # Coleta de informações do usuário
    nome = input("Olá! Qual o seu nome? ")
    idade = int(input("Qual a sua idade? "))
    altura = float(input("Qual a sua altura em metros? "))
    peso = float(input("Qual o seu peso em quilogramas? "))

    # Pergunta sobre o objetivo do usuário
    print("\nQual o seu principal objetivo?")
    print("1. Hipertrofia")
    print("2. Definição muscular")
    print("3. Saúde e bem-estar")
    objetivo_num = int(input("Digite o número da sua opção: "))

    if objetivo_num == 1:
        objetivo = "Hipertrofia"
    elif objetivo_num == 2:
        objetivo = "Definição muscular"
    elif objetivo_num == 3:
        objetivo = "Saúde e bem-estar"
    else:
        print("Opção inválida. Assumindo 'Saúde e bem-estar' como objetivo.")
        objetivo = "Saúde e bem-estar"

    # Pergunta sobre a experiência do usuário com treino
    print("\nQual a sua experiência com treino?")
    print("1. Nunca treinei antes")
    print("2. 1 a 3 meses")
    print("3. 6 a 1 ano")
    print("4. Já treinei alguns meses antes, mas estou voltando agora")
    experiencia_num = int(input("Digite o número da sua opção: "))

    if experiencia_num == 1:
        nivel_experiencia = "iniciante (nunca treinou)"
        min_exercicios = 4  # Mínimo de 4 exercícios para iniciantes
    elif experiencia_num == 2:
        nivel_experiencia = "iniciante (1 a 3 meses)"
        min_exercicios = 4  # Mínimo de 4 exercícios para iniciantes
    elif experiencia_num == 3:
        nivel_experiencia = "intermediário (6 a 1 ano)"
        min_exercicios = 5  # Mínimo de 5 exercícios para intermediários
    elif experiencia_num == 4:
        nivel_experiencia = "intermediário (voltando a treinar)"
        min_exercicios = 5  # Mínimo de 5 exercícios para intermediários
    else:
        print("Opção inválida. Assumindo 'Iniciante (nunca treinou)' como nível de experiência.")
        nivel_experiencia = "iniciante (nunca treinou)"
        min_exercicios = 4  # Mínimo de 4 exercícios para iniciantes

    # Pergunta sobre o tipo corporal do usuário
    print("\nQual o seu tipo corporal?")
    print("1. Ectomorfo")
    print("2. Mesomorfo")
    print("3. Endomorfo")
    print("4. Não sei, me explique")
    tipo_corporal_num = int(input("Digite o número da sua opção: "))

    if tipo_corporal_num == 1:
        tipo_corporal = "Ectomorfo"
    elif tipo_corporal_num == 2:
        tipo_corporal = "Mesomorfo"
    elif tipo_corporal_num == 3:
        tipo_corporal = "Endomorfo"
    elif tipo_corporal_num == 4:
        print("\nPara descobrir seu tipo corporal, siga estes passos:")
        print("1. **Envolva seu pulso com o polegar e o dedo médio da sua mão dominante.**")
        print("2. **Observe a posição dos seus dedos:**")
        print("   * **Ectomorfo:** Se o polegar e o dedo médio se sobrepõem com facilidade.")
        print("   * **Mesomorfo:** Se o polegar e o dedo médio se tocam.")
        print("   * **Endomorfo:** Se o polegar e o dedo médio não se tocam.")
        tipo_corporal = input("\nAgora, qual você acha que é o seu tipo corporal? ")
    else:
        print("Opção inválida. O tipo corporal não será considerado na geração do treino.")
        tipo_corporal = ""

    # Pergunta quantos dias o usuário quer treinar
    print("\nQuantos dias por semana você quer treinar?")
    print("1. 2 vezes")
    print("2. 3 vezes")
    print("3. 4 vezes")
    print("4. 5 vezes")
    print("5. 6 vezes")
    print("⚠️  Lembre-se que é importante ter pelo menos um dia de descanso para recuperação muscular. Treinar todos os dias pode ser prejudicial.")
    dias_treino_num = int(input("Digite o número da sua opção: "))

    if dias_treino_num == 1:
        dias_treino = "2 vezes"
    elif dias_treino_num == 2:
        dias_treino = "3 vezes"
    elif dias_treino_num == 3:
        dias_treino = "4 vezes"
    elif dias_treino_num == 4:
        dias_treino = "5 vezes"
    elif dias_treino_num == 5:
        dias_treino = "6 vezes"
    else:
        print("Opção inválida. Assumindo 3 vezes por semana.")
        dias_treino = "3 vezes"

    # Pergunta em quanto tempo o usuário quer atingir o objetivo
    tempo_objetivo = input("\nEm quanto tempo você gostaria de atingir seu objetivo (ex: 3 meses, 1 ano)? ")

    # Cálculo do IMC (Índice de Massa Corporal)
    imc = calcular_imc(peso, altura)

    # Imprime os resultados e recomendações
    print(f"\n{nome}, seu IMC é: {imc:.2f}")

    if imc < 18.5:
        print("Você está abaixo do peso ideal. É recomendável consultar um nutricionista.")
    elif 18.5 <= imc < 24.9:
        print("Parabéns! Você está dentro da faixa de peso ideal.")
    elif 25 <= imc < 29.9:
        print("Você está com sobrepeso. É recomendável adotar hábitos alimentares mais saudáveis e praticar exercícios físicos regularmente.")
    else:
        print("Você está obeso. É fundamental procurar ajuda médica para desenvolver um plano de emagrecimento seguro e eficaz.")

    # Configurações de geração e segurança
    generation_config = {
        "candidate_count": 1,
        "temperature": 0.5,
        "top_p": 0.5,
        "top_k": 50,
    }

    safety_settings = {
        'HARASSMENT': 'BLOCK_LOW_AND_ABOVE',
        "HATE": 'BLOCK_LOW_AND_ABOVE',
        "SEXUAL": 'BLOCK_LOW_AND_ABOVE',
        "DANGEROUS": 'BLOCK_LOW_AND_ABOVE',
    }

    # Cria o modelo GenerativeModel
    model = genai.GenerativeModel(
        model_name="gemini-1.0-pro",
        generation_config=generation_config,
        safety_settings=safety_settings
    )

    # Gera sugestões de treino usando o modelo
    sugestoes_treino_texto = gerar_sugestoes_treino(
        model,  # Passa o modelo como argumento
        objetivo,
        imc,
        nivel_experiencia,
        idade,
        tipo_corporal,
        dias_treino,
        tempo_objetivo,
        min_exercicios  # Passa o número mínimo de exercícios
    )

    # Imprime o título do plano de treino
    print(f"\n## Plano de Treino para {objetivo} - Meta: {tempo_objetivo}\n")

    # Processa e imprime as tabelas de treino para cada dia, garantindo separação clara
    for dia, content_item in enumerate(sugestoes_treino_texto, 1):
        texto_treino = content_item.text
        linhas = texto_treino.strip().split("\n")
        tabela_treino = []
        for linha in linhas:
            if "|" in linha:
                colunas = linha.split("|")
                # Substitui valores vazios por "---"
                colunas = ["---" if coluna.strip() == "" else coluna.strip() for coluna in colunas]
                tabela_treino.append(colunas)

        print(f"\n### Dia {dia}\n")  # Título do dia com formatação mais clara
        print(tabulate(tabela_treino, headers="firstrow", tablefmt="grid"))
        print("\n")  # Adiciona uma linha em branco entre as tabelas

def calcular_imc(peso, altura):
    """Calcula o IMC (Índice de Massa Corporal)."""
    return peso / (altura ** 2)

def gerar_sugestoes_treino(
    model, # Recebe o modelo como argumento
    objetivo,
    imc,
    nivel_experiencia,
    idade,
    tipo_corporal,
    dias_treino,
    tempo_objetivo,
    min_exercicios  # Recebe o número mínimo de exercícios
):
    """Gera sugestões de treino usando o modelo GenerativeModel."""

    prompt = f"""
    Crie um plano de treino para alguém com {idade} anos, o objetivo de '{objetivo}', IMC de {imc:.2f},
    nível de experiência '{nivel_experiencia}', tipo corporal '{tipo_corporal}' e que deseja treinar {dias_treino} por semana.
    O objetivo deve ser atingido em {tempo_objetivo}.

    Leve em consideração:
    * Se for adolescente (13 a 15 anos), diminua a intensidade e use exercícios mais práticos, pois os músculos estão em desenvolvimento.
    * Se for adulto, pode usar exercícios mais comuns e eficientes.
    * Se for idoso, sugira exercícios com menos risco de lesão, levando em consideração a possibilidade de fragilidade articular, óssea e muscular.
    * Adapte o treino ao tipo corporal, considerando as características de cada um (Ectomorfo, Mesomorfo, Endomorfo).
    * Inclua pelo menos {min_exercicios} exercícios de treino por dia, mesmo que seja necessário diminuir a intensidade
      para alguns deles, a fim de garantir variedade no treino.

    Para cada dia de treino, inclua:
    * **Alongamento:** Pelo menos dois exercícios de alongamento específicos para o grupo muscular que será treinado no dia.
    * **Treino:**
        * Tipo de treino (ex: cardio, musculação, yoga)
        * Exemplos de exercícios específicos para cada dia de treino, com:
            * Carga (ex: 5kg, 10kg)
            * Número de séries (ex: 3 séries)
            * Repetições (ex: 10 repetições)
            * Intervalo de descanso entre as séries (ex: 60 segundos)
    * Se o nível de experiência for iniciante, priorize treinos de adaptação com cargas mais leves, menos repetições
        e intervalos de descanso maiores.
    * Para pessoas que estão voltando a treinar, sugira um retorno gradual às atividades com cargas leves, menos repetições
        e intervalos de descanso maiores.
    * Certifique-se de que o plano seja desafiador, mas seguro para o usuário, levando em conta sua idade, experiência e tipo corporal.

    Apresente o plano de treino em formato de tabela, separando cada dia de treino com um espaçamento claro:

    ### Dia 1

    Dia 1 | Exercício | Séries | Repetições | Carga | Descanso
    ------- | -------- | -------- | -------- | -------- | --------
    Alongamento | Alongamento 1 | 1 | 30 segundos | --- | 15 segundos
    Alongamento | Alongamento 2 | 1 | 30 segundos | --- | 15 segundos
    Treino | Exercício 1 | 3 | 12 | 10kg | 60 segundos
    Treino | Exercício 2 | 3 | 10 | 8kg | 60 segundos
    Treino | Exercício 3 | 3 | 15 | --- | 45 segundos

    -------------------------------------------------------------------

    ### Dia 2

    Dia 2 | Exercício | Séries | Repetições | Carga | Descanso
    ------- | -------- | -------- | -------- | -------- | --------
    Alongamento | Alongamento 1 | 1 | 30 segundos | --- | 15 segundos
    Alongamento | Alongamento 2 | 1 | 30 segundos | --- | 15 segundos
    Treino | Exercício 1 | 3 | 12 | 10kg | 60 segundos
    Treino | Exercício 2 | 3 | 10 | 8kg | 60 segundos
    Treino | Exercício 3 | 3 | 15 | --- | 45 segundos

    ### Dia 3

     Dia 3 | Exercício | Séries | Repetições | Carga | Descanso
    ------- | -------- | -------- | -------- | -------- | --------
    Alongamento | Alongamento 1 | 1 | 30 segundos | --- | 15 segundos
    Alongamento | Alongamento 2 | 1 | 30 segundos | --- | 15 segundos
    Treino | Exercício 1 | 3 | 12 | 10kg | 60 segundos
    Treino | Exercício 2 | 3 | 10 | 8kg | 60 segundos
    Treino | Exercício 3 | 3 | 15 | --- | 45 segundos

    -------------------------------------------------------------------

    siga esse exemplo de forma exata

    ...
    """

    response = model.generate_content(prompt)  # Usa model.generate_content

    return response  # retorna a resposta do modelo

# Chama a função principal
training_generator()
```
